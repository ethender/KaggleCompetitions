```python
import numpy as np
import pandas as pd
```

```python
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
train_data.head()
```

```python
test_data.head()
```

```python
y = train_data['label']
train_data.drop(['label'],inplace=True,axis=1)
```

```python
train_data.head()
```

```python
import matplotlib as mat
import matplotlib.pyplot as plt

first_digit = train_data.iloc[3,:]
first_digit = first_digit.values.reshape(28,28)

plt.imshow(first_digit,cmap=mat.cm.binary)
plt.axis("off")
```

```python
from sklearn.model_selection import train_test_split

x_train,x_validate,y_train,y_validate = train_test_split(train_data,y,test_size=0.3,random_state=42)
```

```python
from sklearn.svm import SVC

svm_clf = SVC(gamma="auto",random_state=42)
svm_clf.fit(x_train,y_train)
```

```python
y_pred = svm_clf.predict(x_validate)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_validate,y_pred)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_validate,y_pred)
```
